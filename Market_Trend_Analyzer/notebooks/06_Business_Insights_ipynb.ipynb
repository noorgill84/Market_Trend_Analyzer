```python
import pandas as pd
import numpy as np

# Load Dashboard Dataset
df = pd.read_csv("Dashboard_Data.csv")

# Convert date columns
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])

df.head()
```

```python
#overall business KPIs
total_sales = df["Sales"].sum()
total_profit = df["Profit"].sum()
total_orders = df["Order ID"].nunique()
total_customers = df["Customer ID"].nunique()
avg_order_value = total_sales / total_orders

print("="*50)
print("BUSINESS KPIs")
print("="*50)

print(f"Total Sales        : ${total_sales:,.2f}")
print(f"Total Profit       : ${total_profit:,.2f}")
print(f"Total Orders       : {total_orders}")
print(f"Total Customers    : {total_customers}")
print(f"Average Order Value: ${avg_order_value:,.2f}")
```

```python
#top selling categories
category_sales = df.groupby("Category")["Sales"].sum().sort_values(ascending=False)

print("\nTop Selling Category")
print(category_sales)

best_category = category_sales.idxmax()

print(f"\nBest Category: {best_category}")
```

```python
#most profitable category
category_profit = df.groupby("Category")["Profit"].sum().sort_values(ascending=False)

print(category_profit)

print(f"\nMost Profitable Category: {category_profit.idxmax()}")
```

```python
#top 10 products
top_products = (
    df.groupby("Product Name")["Sales"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

print(top_products)
```

```python
#best region
region_sales = df.groupby("Region")["Sales"].sum().sort_values(ascending=False)

print(region_sales)

print(f"\nBest Region: {region_sales.idxmax()}")
```

```python
#best state
state_sales = (
    df.groupby("State")["Sales"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

print(state_sales)
```

```python
#best customer segment
segment_sales = (
    df.groupby("Segment")["Sales"]
      .sum()
      .sort_values(ascending=False)
)

print(segment_sales)

print(f"\nBest Customer Segment: {segment_sales.idxmax()}")
```

```python
#monthly sales analysis
monthly_sales = (
    df.groupby(df["Order Date"].dt.month_name())["Sales"]
      .sum()
)

months = [
    "January","February","March","April","May","June",
    "July","August","September","October","November","December"
]

monthly_sales = monthly_sales.reindex(months)

print(monthly_sales)
```

```python
#highest sales month
best_month = monthly_sales.idxmax()

print("Highest Sales Month:", best_month)
```

```python
monthly_profit = (
    df.groupby(df["Order Date"].dt.month_name())["Profit"]
      .sum()
)

monthly_profit = monthly_profit.reindex(months)

print(monthly_profit)
```

```python
#highest profit month
print("Highest Profit Month:", monthly_profit.idxmax())
```

```python
#shipping performance
df["Shipping Days"] = (
    df["Ship Date"] - df["Order Date"]
).dt.days

print(df["Shipping Days"].describe())

print(f"\nAverage Shipping Time: {df['Shipping Days'].mean():.2f} Days")
```

```python
#discount analysis
discount_profit = (
    df.groupby("Discount")["Profit"]
      .mean()
)

print(discount_profit)
```

```python
#top 10 customers
customers = (
    df.groupby("Customer Name")["Sales"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

print(customers)
```

```python
#market trend score
monthly = (
    df.groupby(df["Order Date"].dt.to_period("M"))["Sales"]
      .sum()
)

monthly = monthly.reset_index()

monthly["Order Date"] = monthly["Order Date"].astype(str)

monthly["Growth %"] = (
    monthly["Sales"].pct_change() * 100
)

monthly.head()
```

```python
#trend classification
def classify(growth):

    if pd.isna(growth):
        return "No Data"

    elif growth > 10:
        return "Growing 📈"

    elif growth < -10:
        return "Declining 📉"

    else:
        return "Stable ➖"

monthly["Trend"] = monthly["Growth %"].apply(classify)

monthly
```

```python
#overall trend
latest = monthly.iloc[-1]

print("="*40)

print("LATEST MARKET STATUS")

print("="*40)

print("Month :", latest["Order Date"])

print("Growth :", round(latest["Growth %"],2),"%")

print("Trend :", latest["Trend"])
```

```python
#business recommendations
print("="*50)
print("BUSINESS RECOMMENDATIONS")
print("="*50)

print(f"✅ Focus on {best_category} products as they generate the highest sales.")

print(f"✅ Increase inventory in the {region_sales.idxmax()} region.")

print(f"✅ Target {segment_sales.idxmax()} customers with marketing campaigns.")

print(f"✅ Prepare additional stock before {best_month}, which has the highest sales.")

if latest["Trend"] == "Growing 📈":
    print("✅ Market demand is increasing. Consider expanding inventory.")

elif latest["Trend"] == "Declining 📉":
    print("⚠️ Sales are declining. Review pricing, promotions, and marketing strategies.")

else:
    print("ℹ️ Sales are relatively stable. Continue monitoring trends.")
```

```python
#save insights
summary = {
    "Total Sales": total_sales,
    "Total Profit": total_profit,
    "Top Category": best_category,
    "Best Region": region_sales.idxmax(),
    "Best Segment": segment_sales.idxmax(),
    "Best Month": best_month,
    "Market Trend": latest["Trend"]
}

summary_df = pd.DataFrame([summary])

summary_df.to_csv("business_summary.csv", index=False)

print("Business Summary Saved Successfully!")
```
