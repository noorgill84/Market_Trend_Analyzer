```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Optional: Better plot style
plt.style.use("ggplot")

# Load cleaned dataset
df = pd.read_csv("Cleaned_Superstore.csv")

# Convert date columns
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
```

```python
print("Total Sales =", round(df["Sales"].sum(),2))
print("Total Profit =", round(df["Profit"].sum(),2))
print("Total Orders =", df["Order ID"].nunique())
```

```python
#sales by category
category_sales = df.groupby("Category")["Sales"].sum().sort_values()

plt.figure(figsize=(8,5))

category_sales.plot(kind="bar")

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")

plt.show()
```

```python
#profit by category
category_profit = df.groupby("Category")["Profit"].sum().sort_values()

plt.figure(figsize=(8,5))

category_profit.plot(kind="bar")

plt.title("Profit by Category")

plt.show()
```

```python
#sales by region
region_sales = df.groupby("Region")["Sales"].sum()

plt.figure(figsize=(8,5))

region_sales.plot(kind="bar")

plt.title("Region-wise Sales")

plt.show()
```

```python
#profits by region
region_profit = df.groupby("Region")["Profit"].sum()

plt.figure(figsize=(8,5))

region_profit.plot(kind="bar")

plt.title("Region-wise Profit")

plt.show()
```

```python
#top 10 products
top_products = df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))

top_products.plot(kind="bar")

plt.title("Top 10 Products by Sales")

plt.xticks(rotation=90)

plt.show()
```

```python
#monthly sales trend
monthly_sales = df.groupby("Month Name")["Sales"].sum()

months = ["January","February","March","April","May","June","July","August","September","October","November","December"]

monthly_sales = monthly_sales.reindex(months)

plt.figure(figsize=(12,5))

plt.plot(monthly_sales.index,
         monthly_sales.values,
         marker='o')

plt.xticks(rotation=45)

plt.title("Monthly Sales Trend")

plt.show()
```

```python
#year wise sales
year_sales = df.groupby("Year")["Sales"].sum()

plt.figure(figsize=(8,5))

plt.plot(year_sales.index,
         year_sales.values,
         marker='o')

plt.title("Year-wise Sales")

plt.show()
```

```python
#segment wise sales
segment_sales = df.groupby("Segment")["Sales"].sum()

plt.figure(figsize=(8,5))

segment_sales.plot(kind="pie",autopct="%1.1f%%")

plt.ylabel("")

plt.title("Customer Segment Distribution")

plt.show()
```

```python
#ship mode analysis
ship = df["Ship Mode"].value_counts()

plt.figure(figsize=(8,5))

ship.plot(kind="bar")

plt.title("Shipping Mode Usage")

plt.show()
```

```python
#discount vs profit
plt.figure(figsize=(8,5))

plt.scatter(df["Discount"],
            df["Profit"])

plt.xlabel("Discount")

plt.ylabel("Profit")

plt.title("Discount vs Profit")

plt.show()
```

```python
#sales ditribtion
plt.figure(figsize=(8,5))

plt.hist(df["Sales"], bins=40)

plt.title("Sales Distribution")

plt.xlabel("Sales")

plt.ylabel("Frequency")

plt.show()
```

```python
#profit distribution
plt.figure(figsize=(8,5))

plt.hist(df["Profit"], bins=40)

plt.title("Profit Distribution")

plt.xlabel("Profit")

plt.show()
```

```python
#top 10 states
state_sales = df.groupby("State")["Sales"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,5))

state_sales.plot(kind="bar")

plt.title("Top 10 States by Sales")

plt.xticks(rotation=45)

plt.show()
```

```python
#top customers
customers = df.groupby("Customer Name")["Sales"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,5))

customers.plot(kind="bar")

plt.title("Top Customers")

plt.xticks(rotation=45)

plt.show()
```

```python
#corre;ation analysis
numeric = df[["Sales","Profit","Quantity","Discount","Shipping Days"]]

corr = numeric.corr()

print(corr)
```
